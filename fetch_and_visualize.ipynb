```python
import pandas as pd
import yfinance as yf

# 1 year history for 5 large cap tickers
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA"]
df = yf.download(tickers, start="2024-07-01", end="2025-07-30")["Close"]

#Quick peek
df.head()
```

```python
# (Optional) Check current working directory
import os
print("Working directory:", os.getcwd())

# Save the closing-price DataFrame into the data/ folder
df.to_csv("data/close_prices.csv", index=True)
print("✅ Saved close_prices.csv to data/")
```

```python
import plotly.express as px

fig = px.line(
    df.reset_index().melt(id_vars="Date", var_name="Ticker", value_name="Close"),
    x="Date", y="Close", color="Ticker",
    title="Closing Prices: AAPL, MSFT, GOOGL, AMZN, TSLA"
)
fig.show()
```
